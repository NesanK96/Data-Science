###### **import necessary libraries**

###### **merge those two csv files (after getting as dataframes, get them as a single dataframe)**

```python
college1 = pd.read_csv('https://raw.githubusercontent.com/NesanK96/Data-Science/main/college_1.csv')

college2 = pd.read_csv('https://raw.githubusercontent.com/NesanK96/Data-Science/main/college_2.csv')

df = college1.append(college2, ignore_index = True)

df
```

###### Take each csv file , split that csv file into multiple categories (example csv files are added in the repo) 


###### **consider if the codekata score exceeds 15000 points(present week) then make a csv on those observations as Exceeded expectations.csv**


```python
Exceeded_expectations = df.loc[df['CodeKata Score'] > 15000]

Exceeded_expectations.to_csv('/content/Exceeded_expectations.csv')

Exceeded_expectations
```

###### **if  10000<codekata score<15000   (Reached_expectations.csv)**



```python
Reached_expectations = df[(df['CodeKata Score'] > 10000) & (df['CodeKata Score']<15000)]

Reached_expectations.to_csv('/content/Reached_expectations.csv')

Reached_expectations
```

###### **if  7000<codekata score<10000   (Needs_Improvement.csv)**


```python
Needs_improvement = df.loc[(df['CodeKata Score']>7000) & (df['CodeKata Score']<10000)]

Needs_improvement.to_csv('/content/Needs_improvement.csv')

Needs_improvement
```

###### **if  codekate score < 7000        (Unsatisfactory.csv)**

```python
Unsatisfactory = df[df['CodeKata Score']<7000]

Unsatisfactory.to_csv('/content/Unsatisfactory.csv')

Unsatisfactory
```

###### **Average of previous week geekions vs this week geekions (i.e Previous Geekions vs CodeKata Score)**

```python
Previous = df['Previous Geekions'].mean()
Current = df['CodeKata Score'].mean()

print('Previous week Geekions : ', Previous, '\nCurrent week Geekions : ', Current)
```

###### **No of students participated** 

```python
Previous = df['Previous Geekions'].count()
Current = df['CodeKata Score'].count()

print('Previous week Participation : ', Previous, '\nCurrent week Participation : ', Current)
```

###### #**Average completion of python course or my_sql or python english or computational thinking**

```python
Python_Course = df['python'].mean()
My_SQL = df['mysql'].mean()
Python_English = df['python_en'].mean()
Computational_Thinking = df['computational_thinking'].mean()


print('Python Course: ', Python_Course, '\nMy_SQL : ', My_SQL, '\nPython English : ', Python_English, '\nComputational Thinking : ', Computational_Thinking)
```

###### **rising star of the week (top 3 candidate who performed well in that particular week)**

```python
Rising_Star = df.sort_values(by = 'Rising')

Rising_Star = Rising_Star[ :-4:-1]

Rising_Star
```

###### **Shining stars of the week (top 3 candidates who has highest geekions)**

```python
Shining_Stars = df.sort_values(by = 'CodeKata Score', ascending = False)

Shining_Stars = Shining_Stars[:3]

Shining_Stars
```

```python
import plotly.express as px

Department = px.pie(df, values = 'CodeKata Score', names = 'Department', title = 'Department wise Codekata performance')

Department.update_traces(textposition = 'inside', textinfo = 'percent+label',
                         insidetextorientation = 'radial')

# Department.update_layout(uniformtext_minsize=10)

Department.show()
```

###### **Department wise toppers (horizantal bar graph or any visual representations of your choice)**

```python
df_new = df.sort_values(['CodeKata Score'], ascending=False).groupby('Department').head(3)

df_new
```

```python
Department_Toppers = px.bar(df_new, x='Department', y = 'CodeKata Score', 
                          color = 'Name', title = 'Department wise Toppers',
                          text_auto=True)

Department_Toppers.show()
```
